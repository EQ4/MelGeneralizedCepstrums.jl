```julia
using MelGeneralizedCepstrums
using WAV
using DSP
using PyPlot
```

```julia
x, fs = wavread(joinpath(Pkg.dir("VoiceConversion"), "test", "data", "test16k.wav"))
x = vec(x)

# Visualize the speech signal in time-domain
figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(x), x, label="a speech signal")
xlim(1, endof(x))
xlabel("sample")
legend()
```

```julia
# Pick a short segment
pos = 3000
fftlen = 1024
xw = x[pos+1:pos+fftlen] .* blackman(fftlen)

figure(figsize=(16, 6), dpi=80, facecolor="w", edgecolor="k")
plot(1:endof(xw), xw, linewidth="2", label="a windowed speech signal")
xlim(1, endof(xw))
xlabel("sample")
legend()
```

```julia
function pplot(sp, envelope; title="envelope")
    figure(figsize=(16, 8), dpi=80, facecolor="w", edgecolor="k")
    plot(sp, "b-", linewidth="2", label="Original log spectrum 20log|X(ω)|")
    plot(20/log(10)*(envelope), "r-", linewidth="3", label=title)
    xlim(1, length(sp))
    xlabel("frequency bin")
    ylabel("log amplitude")
    legend()
end
```

```julia
# Compute spectrum 20log|X(ω)| for a windowed signal
sp = 20log10(abs(rfft(xw)));
```

```julia
# Linear Cepstrum
c = mcep(xw, 20, 0.0)
pplot(sp, mgc2sp(c, fftlen), title="Linear cepstrum based envelope")
```

```julia
# Mel-Cepstrum
mc = mcep(xw, 20, 0.41)
pplot(sp, mgc2sp(mc, fftlen), title="Mel-cepstrum based envelope")
```

```julia
# Linear Prediction
mgc = mgcep(xw, 20, 0.0, -1.0)
pplot(sp, mgc2sp(mgc, fftlen), title="Linear Prediction based envelope")
```

```julia
# Warped Linear Prediction
mgc = mgcep(xw, 20, 0.41, -1.0)
pplot(sp, mgc2sp(mgc, fftlen), title="Warped Linear Prediction based envelope")
```

```julia
# Generalized Cepstrum
mgc = mgcep(xw, 20, 0.0, -0.35)
pplot(sp, mgc2sp(mgc, fftlen), title="Generalized Cepstrum based envelope")
```

```julia
# Mel-Generalized Cepstrum
mgc = mgcep(xw, 20, 0.41, -0.35)
pplot(sp, mgc2sp(mgc, fftlen), title="Mel-Generalized Cepstrum based envelope")
```
